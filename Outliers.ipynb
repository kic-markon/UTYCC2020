```python
# https://ocefpaf.github.io/python4oceanographers/blog/2015/03/16/outlier_detection/
from pandas import read_table
import matplotlib.pyplot as plt

# Read in the data file 
fname = 'soil.dat'
cols = ['t', 'u']
df = read_table(fname , delim_whitespace=True, names=cols)

# Plot the raw data
plt.plot(df['t'],df['u'])
plt.show()
```

```python
# Get the median filter from scipy
from scipy.signal import medfilt
df['m'] = medfilt(df['u'])

# Each data point is now replaced with the median
plt.plot(df['t'],df['m'])
plt.show()
```
